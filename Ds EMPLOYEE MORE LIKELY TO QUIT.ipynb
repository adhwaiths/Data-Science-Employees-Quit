```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_cat = onehotencoder.fit_transform(X_cat).toarray()
X_cat.shape
```

```python
X_cat=pd.DataFrame(X_cat)
```

```python
X_cat
#convert 1470x6 to 1470x26 i.e encoded every single column value to binary value
```

```python
X_numerical = employee_df[['Age', 'DailyRate', 'DistanceFromHome',	'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement',	'JobLevel',	'JobSatisfaction',	'MonthlyIncome',	'MonthlyRate',	'NumCompaniesWorked',	'OverTime',	'PercentSalaryHike', 'PerformanceRating',	'RelationshipSatisfaction',	'StockOptionLevel',	'TotalWorkingYears'	,'TrainingTimesLastYear'	, 'WorkLifeBalance',	'YearsAtCompany'	,'YearsInCurrentRole', 'YearsSinceLastPromotion',	'YearsWithCurrManager']]
X_numerical
#attrition was not included since its the feature that has to be predicted. Only coulmns that were already numerical were included.
```

```python
X_all = pd.concat([X_cat, X_numerical], axis = 1)
X_all.shape
```

```python
X_all.columns=X_all.columns.astype(str)
#convert all columns to string names and not int and str(some columns from onehotencoder had integer column names)
```

```python
X_all.columns.dtype
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X_all)
```

```python
y = employee_df['Attrition']
```

```python
y
```

```python
#creating training set and testing set data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#training the logistiregression classifier
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(x_train,y_train)
```

```python
#predicting using testing set
pred = lg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
#evaluating performance of the trained model
print("Accuracy {} %".format( 100 * accuracy_score(pred, y_test)))
```

```python
cm = confusion_matrix(pred, y_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, pred))
```

```python
#lets try another ML algorithm RandomForestClassifer
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train, y_train)
prediction=rf.predict(x_test)
```

```python
# Evaluating the model Performance
cm = confusion_matrix(prediction, y_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, prediction))
```
